```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

EDA

```python
df = pd.read_csv("./ACI-IoT-2023.csv")
```

```python
pd.set_option('display.max_columns', None)
df.describe(include='all')
```

```python
df.shape
```

```python
df.drop(["Flow Bytes/s","Timestamp","Flow Packets/s"],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df.Label.value_counts() 
```

```python
label_counts = df.Label.value_counts()
```

```python
filtered_label_counts = label_counts[label_counts.index != 'Benign']

hatch_patterns = ['/', '\\', '|', '-', '+', 'x', 'o', '*', '.']

plt.figure(figsize=(20, 8))
ax = sns.barplot(x=filtered_label_counts.index, y=filtered_label_counts.values, palette="Blues")

for i, bar in enumerate(ax.patches):
    bar.set_hatch(hatch_patterns[i % len(hatch_patterns)])  

plt.title('Anomaly Counts', fontsize=20)
plt.xlabel('Labels', fontsize=20)
plt.ylabel('Count', fontsize=20)
plt.xticks(fontsize=16, rotation=45)
plt.yticks(fontsize=16)

plt.show()
```

Label Encoder

```python
df['Label'] = df['Label'].apply(lambda x: 0 if x == 'Benign' else 1)
```

```python
df.Label.value_counts()
```

```python
data = {
    'Category': ['Normal Data', 'Anomaly'],
    'Count': [329295, 902116]
}

d = pd.DataFrame(data)

hatch_patterns = ['/', '\\']  

plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Category', y='Count', data=d, palette="Blues")

for i, bar in enumerate(ax.patches):
    bar.set_hatch(hatch_patterns[i % len(hatch_patterns)])

plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Normal Data vs Anomaly (Before Over sampling)', fontsize=16)

plt.show()
```

Normal Data Oversampling

```python
benign_samples = df[df['Label'] == 0]
num_to_add =  1000000 - len(benign_samples)

# Randomly sample from existing benign data
synthetic_samples = benign_samples.sample(n=num_to_add, replace=True, random_state=42)

# Append to the original dataset
df_balanced = pd.concat([df, synthetic_samples], ignore_index=True)

print(df_balanced['Label'].value_counts())
```

```python
data = {
    'Category': ['Normal Data', 'Anomaly'],
    'Count': [1000000, 902116]
}

d = pd.DataFrame(data)

hatch_patterns = ['/', '\\']  

plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Category', y='Count', data=d, palette="Blues")

for i, bar in enumerate(ax.patches):
    bar.set_hatch(hatch_patterns[i % len(hatch_patterns)])

plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Normal Data vs Anomaly (After Oversampling)', fontsize=16)

plt.show()
```

```python
df_balanced.to_csv('balanced_dataset.csv', index=False)
print("Dataset saved successfully as 'balanced_dataset.csv'")
```
